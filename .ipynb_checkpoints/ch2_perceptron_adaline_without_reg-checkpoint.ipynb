```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
class Perceptron:
    def __init__(self, eta=0.01, n_iter=10): # eta : learning rate
        self.eta    = eta
        self.n_iter = n_iter
    
    def fit(self, X, y): # X: data matrix, y: target values
        n_rows, n_factors = X.shape
        self.weights = np.zeros(n_factors + 1)
        self.errors = [] # keep track of error count per iter
        
        for i in range(self.n_iter):
            error_count = 0 
            for row, target in zip(X, y): # zip matches each row with the target_value
                update = self.eta * (target - self.predict(row)) # when the prediction and target match, the elem in update is 0; otherwise update causes the weights to move in the opposite direction of the prediction                
                self.weights += update * np.insert(row, 0, 1) # add 1 to the front of row, so that the constant also gets updated # syntax: np.insert(array, index, same_type_values_to_be_inserted) => vector (matrix will be flattened)
                error_count += int(update != 0) # increment error count if prediction is wrong
            self.errors.append(error_count)
        return self
    
    # return 1 or -1
    def predict(self, row):
        return np.where(self.net_input(row) >= 0, 1, -1) # syntax: np.where(cond, value_when_true, value_when_false) => array; if net > 0, assign positive class
    
    # returns a float value
    def net_input(self, row):
        return np.dot(row, self.weights[1:]) + self.weights[0] # dot product plus constant
        # note: when np.dot or array.dot is applied, the second matrix is automaticaly transposed if the dimensions do not match
        
```

```python
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, resolution = 0.02):
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))]) # num of unique classes
    
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1 # min & max value of att1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1 # min & max value of att2
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                     np.arange(x2_min, x2_max, resolution))  # np.arrange generates array with evenly space intervals np.arange(min, max, interval)
    dots = np.array([xx1.ravel(), xx2.ravel()]).T # Note: ravel does not copy the values, thus update will alter the origial matrix, but its performace is better than flatten
    # with meshgrid, ravel and transforms, coords of a rectangular set of dots are generated

    Z = classifier.predict(dots) # mark each dot as 1 or -1 depending on the prediction
    Z = Z.reshape(xx1.shape) # since the dots array are flattened, this reshapes it make to a matrix of dots
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    # plot class samples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                    alpha=0.8, c=cmap(idx),
                    marker=markers[idx], label=cl)
```

```python
plot_decision_regions(X, y, classifier=ppn)
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')
plt.show()
```

```python
class AdalineGD: # the original code write AdalineGD(object) which in syntax for python 2, in python 3, all classes inherit from Object
    def __init__(self, eta=0.01, n_iter=50):
        self.eta = eta
        self.n_iter = n_iter
    
    def fit(self, X, y):
        n_rows, n_factors = X.shape
        self.weights = np.zeros(n_factors + 1)
        self.costs = [] # keep track of cost per iter
        
        for i in range(self.n_iter):
            output = self.net_input(X) # all_predictions
            errors = y - output # an array of all prediction vs target results
            # instead of looping through each row, update weight with result of all rows
            self.weights[1:] += self.eta * X.T.dot(errors) 
            self.weights[0] += self.eta * errors.sum()
            cost = (errors ** 2).sum() / 2.0 # cost fuction: sum of square means / 2; note: I believe the divison by 2 is not a must, it just allows the differentiation function to not carry any multiple
            self.costs.append(cost)
        return self
    
    # returns a float value
    def net_input(self, X):
        return np.dot(X, self.weights[1:]) + self.weights[0] # dot product plus constant
    
    # for predicting a test set
    def predict(self, X):
        return np.where(self.activation(X) >= 0.0, 1, -1)
    
    def activation(self, X):   
        return self.net_input(X)
```

```python
ada1 = AdalineGD(n_iter=10, eta=0.01).fit(X,y)
ada2 = AdalineGD(n_iter=10, eta=0.0001).fit(X,y)
```

```python
def standardize(X):
    return (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
X_std = standardize(X)
ada1 = AdalineGD(n_iter=10, eta=0.01).fit(X_std,y)
ada2 = AdalineGD(n_iter=10, eta=0.0001).fit(X_std,y)
for context, cost in { 'learning rate 0.01': ada1.costs, 'learning rate 0,0001': ada2.costs }.items():
    plt.figure()
    f = sns.lineplot(range(1, len(cost)+1), cost)
    if max(cost) > (10 ** 10):
        f.set_yscale('log') 
    plt.title(context)
    plt.xlabel('iters')
    plt.ylabel('errors')
```

```python
from numpy.random import seed

class AdalineSGD: # stochastic/on-line gradient descent
    def __init__(self, eta=0.01, n_iter=10, shuffle=True, random_state=None): # random state is for a providing a fixed seed
        self.eta = eta
        self.n_iter = n_iter
        self.shuffle = shuffle
        if random_state:
            seed(random_state)
    
    def fit(self, X, y):
        n_rows, n_factors = X.shape
        self.weights = self._init_weights(n_factors)
        self.costs = [] # keep track of cost per iter
        
        for i in range(self.n_iter):
            self.costs_in_curr_iter = []
            if self.shuffle:
                X, y = self._shuffle(X, y) # Note: X defined in window scope is unchanged
            
            for row, target in zip(X, y):
                cost = self._update_weights_and_costs(row, target)  
                self.costs_in_curr_iter.append(cost) 
            avg_cost = np.array(self.costs_in_curr_iter).mean()    
            self.costs.append(avg_cost) 
        return self
    
    # for on-line updating weight with new data
    def partial_fit(self, X, y):
        self.weights = self._init_weights(X.shape[1]) # only init if not exist
        
        # force y to be an array
        if type(y) == int:
            y = np.array([y])
            
        for row, target in zip(X, y):
            self._update_weights_and_costs(row, target)    
        return self
        
            
    def _init_weights(self, n_factors):
        try:
            self.weights
        except AttributeError:
            self.weights = np.zeros(n_factors + 1)
        return self.weights
    
    def _shuffle(self, X, y):
        rand_order = np.random.permutation(len(y)) # generates an array of shuffled vales from 0 to len(y)
        return X[rand_order], y[rand_order]
    
    def _update_weights_and_costs(self, row, target):
        error = target - self.net_input(row)
        self.weights += self.eta * error * np.insert(row, 0, 1)
        return 0.5 * error ** 2
    
    # returns a float value
    def net_input(self, row):
        return np.dot(row, self.weights[1:]) + self.weights[0] 
    
    # for predicting test set
    def predict(self, X):
        return np.where(self.activation(X) >= 0.0, 1, -1)

    def activation(self, X):   
        return self.net_input(X)
       
```

```python
ada = AdalineSGD(n_iter=15, eta=0.01, random_state=1)
ada.fit(X_std,y)
```

```python
sns.lineplot(range(1, len(ada.costs) + 1), ada.costs, marker='o', color='red', size=100)
plt.xlabel('epochs')
plt.ylabel('avg cost')
```

```python
plot_decision_regions(X_std, y, classifier=ada)
plt.xlabel('sepal length [std]')
plt.ylabel('petal length [std]')
plt.legend(loc='upper left')
plt.show()
```

```python
ada = AdalineSGD(n_iter=15, eta=0.01, random_state=1)
ada.partial_fit(X_std[:50],y[:50])
ada.partial_fit(X_std[50:],y[50:])

plot_decision_regions(X_std, y, classifier=ada)
plt.xlabel('sepal length [std]')
plt.ylabel('petal length [std]')
plt.legend(loc='upper left')
plt.show()
```
