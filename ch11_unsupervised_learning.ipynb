```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(
    n_samples=150,
    n_features=2,
    centers=3,
    cluster_std=0.5,
    shuffle=True,
    random_state=0
)
plt.scatter(X[:,0], X[:,1], c='blue', marker='o',s=50)
plt.grid()
plt.show()
```

```python
# K-means clustering algorithm
from sklearn.cluster import KMeans
for j in range(6):
    n_clusters = j + 1
    km = KMeans(
        n_clusters=n_clusters, # desired num of clusters
    #     init='random', # by default, use K-means++ algo, which improve the seed centroids selected
        n_init=10, # run algo 10 times with different centroids to choose the final model as the one with the lowest SSE
        max_iter=300, # maximum number of iterations for each single run
        tol=1e-04, # tolerance to the change in SSE to declare convergence
        random_state=0
    )
    y_km = km.fit_predict(X)

    clrs = ['r', 'g', 'lightblue', 'orange', 'pink', 'purple']
    plt.figure()
    for i, c in zip(range(n_clusters), clrs):
        plt.scatter(X[y_km == i][:,0], X[y_km == i][:,1], c=c, marker='o', s=50, label='cluster {}'.format(i))
    plt.scatter(
        km.cluster_centers_[:,0],
        km.cluster_centers_[:,1],
        s=250,
        marker='*',
        c='black',
        label='centroids'
    )
    plt.legend()
```

```python
# Note: Fuzzy/Soft Clustering  e.g. fuzzy C-means (FCM)/soft k-means algorithm -> not available in sk-learn
```

```python
# finding the optimal number of clusters
# Note: 'inertia_' = value of the cost function of K-means, i.e. sum of squared distances to centroids

distortions = []
for i in range(1, 11):
    km = KMeans(
        n_clusters=i,
        init='k-means++',
        n_init=10,
        max_iter=300,
        random_state=0
    )
    km.fit(X)
    distortions.append(km.inertia_)

plt.plot(range(1,11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
# elbow at k=3 => distortion decrease very slowly after exceeding k=3
```

```python
# Evaluating clustering by Silhouette Coefficient
# each point's coefficient =avg dist to pts in closest cluster(b) - avg dist to pts in same cluster(a) / max(a,b) 
# Closer to 1 indicates better clustering

from matplotlib import cm # colormap
from sklearn.metrics import silhouette_samples

# example for n_clusters 2, 3 & 4
for i in range(3):
    n_cluster = i + 2
    km = KMeans(
        n_clusters=n_cluster,
        init='k-means++',
        n_init=10, 
        max_iter=300, 
        tol=1e-04, 
        random_state=0
    )
    y_km = km.fit_predict(X)
    
    # scatter plot same as above, for reference
    plt.figure()
    for i in range(n_clusters):
        plt.scatter(X[y_km == i][:,0], X[y_km == i][:,1], color = cm.jet(i / n_clusters), marker='o', s=50, label='cluster {}'.format(i))
    plt.scatter(
        km.cluster_centers_[:,0],
        km.cluster_centers_[:,1],
        s=250,
        marker='*',
        c='black',
        label='centroids'
    )
    plt.legend()
    
    # Silhouette coefficient plot
    cluster_labels = np.unique(y_km)
    silhouette_vals = silhouette_samples(X, y_km, metric='euclidean') # silhouette coeff of each point
    y_ax_lower, y_ax_upper = 0, 0
    yticks = []
    
    plt.figure()
    for i, c in enumerate(cluster_labels):
        c_silhouette_vals = silhouette_vals[y_km == c]
        c_silhouette_vals.sort()
        y_ax_upper += len(c_silhouette_vals)
        color = cm.jet(i / n_clusters) # the jet color set, returns rgba when provided a float

        # horizontal bar plot
        plt.barh(              
          range(y_ax_lower, y_ax_upper),
          c_silhouette_vals,
          height=1.0,
          edgecolor='None',
          color=color
        )
        yticks.append((y_ax_lower + y_ax_upper) / 2)
        y_ax_lower += len(c_silhouette_vals)
    silhouette_avg = np.mean(silhouette_vals)
    plt.axvline(silhouette_avg, c='r', ls='--')
    plt.yticks(yticks, cluster_labels + 1) # reseting the y-axis ticks to represent the cluster numbering
    plt.ylabel('Cluster')
    plt.xlabel('Silhouette Coefficient')
```

```python
# Hierarchical Clustering - Agglomerative hierarchical clustering(iteratively merging clusters) using complete linkage (max distance of pts between clusters)

# example with 5 data points
np.random.seed(123)
variables = ['X', 'Y', 'Z']
labels = ['ID_0','ID_1','ID_2','ID_3','ID_4']
X = np.random.random_sample([5,3])*10
df = pd.DataFrame(X, columns=variables, index=labels)
df
```

```python
# calc distance betweem each point (each point is considered a cluster initially)
from scipy.spatial.distance import pdist, squareform # Note: this came up in ch5 as well
row_dist = pd.DataFrame(squareform(pdist(df, metric='euclidean')), columns=labels, index=labels)
row_dist
```

```python
from scipy.cluster.hierarchy import linkage
# row_clusters = linkage(df.values, method='complete', metric='euclidean')
row_clusters = linkage(pdist(df, metric='euclidean'), method='complete')
```

```python
pd.DataFrame(
    row_clusters,  
    columns=['row label 1','row label 2','distance','no. of items in clust.'],
    index=['cluster %d' %(i+1) for i in range(row_clusters.shape[0])]
)
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import set_link_color_palette
# set_link_color_palette(['black'])

row_dendr = dendrogram(
    row_clusters,
    labels=labels,
#     color_threshold=np.inf
)
plt.tight_layout()
plt.ylabel('Euclidean distance')
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
axd = fig.add_axes([0.09,0.1,0.2,0.6])
row_dendr = dendrogram(row_clusters, orientation='left')

# heatmap represents dstance
df_rowclust = df.iloc[row_dendr['leaves'][::-1]] # row_dendr['leaves'][::-1] = [4,0,3,2,1]; reorder data in order of leaves
axm = fig.add_axes([0.23,0.1,0.6,0.6])
# plotting the values of the data points as a heatmap to compare more easily their similarity
cax = axm.matshow(df_rowclust, interpolation='nearest', cmap='hot_r')

axd.set_xticks([])
axd.set_yticks([])
for i in axd.spines.values():
    i.set_visible(False)
fig.colorbar(cax)
axm.set_xticklabels([''] + list(df_rowclust.columns))
axm.set_yticklabels([''] + list(df_rowclust.index))
plt.show()
```

```python
# DBSCAN - Density-based Spatial Clustering of Applications with Noise 

# example that cannot be separated by spherical regions, such as using kMeans
from sklearn.datasets import make_moons
X, y = make_moons(
    n_samples=200,
    noise=0.05,
    random_state=0
)
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
from matplotlib import cm # colormap

def plot_graph(clf):
    pred = clf.fit_predict(X)
    plt.figure()
    n_clusters = len(set(pred))
    for i in range(n_clusters):
        plt.scatter(
          X[pred==i,0],
          X[pred==i,1],
          c=cm.jet(i/n_clusters),
          marker='o',
          s=40,
          label='cluster {}'.format(i)
        )
        plt.legend()

km = KMeans(n_clusters=2, random_state=0)
plot_graph(km)
plt.title('K-means clustering')

ac = AgglomerativeClustering(
  n_clusters=2,
  affinity='euclidean',
  linkage='complete'
)
plot_graph(ac)
plt.title('Agglomerative clustering')

from sklearn.cluster import DBSCAN
# note: it is critical to optimize eps and min_samples to achieve an optimal result
db = DBSCAN(
    eps=0.3, # max dist between points in a cluster, e.g. if set to 0.5 - only 1 cluster will be resulted since the 2 shapes differ by >0.3 but <0.5 at the closest
    min_samples=5, # min points within radius needed to form a cluster
    metric='euclidean'
)
plot_graph(db)
plt.title('DBSCAN')
```

```python
# for reference, DBSCAN that is not optimized
db = DBSCAN(
    eps=0.1, # max dist between points in a cluster, e.g. if set to 0.5 - only 1 cluster will be resulted since the 2 shapes differ by >0.3 but <0.5 at the closest
    min_samples=3, # min points within radius needed to form a cluster
    metric='euclidean'
)
plot_graph(db)
plt.title('DBSCAN')
```
